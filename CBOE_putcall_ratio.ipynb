```python
import csv
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from io import BytesIO
```

```python
#get 2013~2016data and 2016~now data
r = requests.get('http://www.cboe.com/publish/scheduledtask/mktdata/datahouse/indexpc.csv')
r2 = requests.get('http://www.cboe.com/publish/scheduledtask/mktdata/datahouse/indexpcarchive.csv')
```

```python
#only reserve Date and P/C Ratio
df =  pd.read_csv(BytesIO(r.content), header=2 , usecols =  ['DATE', 'P/C Ratio'], parse_dates=["DATE"])
df2 = pd.read_csv(BytesIO(r2.content),  header=2 , usecols =  ['Trade_date', 'P/C Ratio'], parse_dates=["Trade_date"])
```

```python
#Rename the column
df.rename(columns={'DATE': 'Date', 'P/C Ratio': 'Value'}, inplace=True)
df2.rename(columns={'Trade_date': 'Date', 'P/C Ratio': 'Value'}, inplace=True)
```

```python
#Merge
result =df2.append(df)
```
